```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name='./'+'airbnb.csv'
df_air=pd.read_csv(file_name)
df_air
```

```python
df1=df_air.copy()
df1.reset_index()
df1.set_index([np.arange(0,3277),df1['host_id']])
df1['total_score']=df1['overall_satisfaction']+df1['reviews']*0.378
df1_1=df1.sort_values('total_score',ascending=True)
df1_1.to_csv('./sorted_total_score_ascend.csv')
df1_2=df1.sort_values('total_score',ascending=False)
df1_2.to_csv('./sorted_total_score_descend.csv')
```

```python
df2=pd.DataFrame()
neighborhood=df_air.copy().groupby(['neighborhood'])
df2['avg of reviews']=neighborhood['reviews'].mean()
df2['avg of overall_satisfaction']=neighborhood['overall_satisfaction'].mean()
df2['avg of price']=neighborhood['price'].mean()
df2['max of reviews']=neighborhood['reviews'].max()
df2['min of reviews']=neighborhood['reviews'].min()
df2['max of price']=neighborhood['price'].max()
df2['min of price']=neighborhood['price'].min()
df2=df2.sort_values('neighborhood',ascending=True)
df2.to_csv('sorted_neighborhood_factors.csv')
df2
```

```python
df3=df_air.copy()
df4=pd.DataFrame()
df3['price']=pd.cut(df3['price'],[0,100,200,300,400,500,1000,5000])
price=df3.groupby('price')
df4['accommodates average']=price['accommodates'].mean()
df4['accommodates median']=price['accommodates'].median()
df4['bedrooms average']=price['bedrooms'].mean()
df4['bedrooms median']=price['bedrooms'].median()
df4['reviews average']=price['reviews'].mean()
df4['reviews nmedian']=price['reviews'].median()
df4['neighbor list']=price['neighborhood'].apply(list)
df4['length']=price['price'].count()
df4.to_csv('sort_ranged_price.csv')
df4
```

```python
df4.plot(y='accommodates average')
plt.show()
plt.savefig('4-1.png')
df4.plot(y='bedrooms average')
plt.show()
plt.savefig('4-2.png')
fig,axes=plt.subplots(nrows=3,ncols=1,sharex=True)
df2.plot(y='avg of reviews',ax=axes[0])
df2.plot(y='avg of overall_satisfaction',ax=axes[1])
df2.plot(y='avg of price',ax=axes[2])
plt.show()
plt.savefig('4-3.png')
```
